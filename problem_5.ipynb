```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

B = 11.1
```

```python
def generate_data(N=10):
    
    np.random.seed(42)
    # Generate N samples uniformly distributed in [0, 1]
    x = np.sort(np.random.uniform(0, 1, N))   
    # Compute y values for sin(2*pi*x)
    y = np.sin(2*np.pi*x)   
    # Generate noise
    noise = np.random.normal(0,1/B)
    
    y_noise = y + noise
    
    return x, y, y_noise
```

```python
x, y, y_noise = generate_data()

# Plot data
plt.plot(x, y, label='y(x)=sin(2πx)')
plt.scatter(x, y_noise, color='red', label='Noise Observations')
plt.xlim([0, 1])
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sinusoidal Function')
plt.show()
```
