```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

__Problem 3__

3.2.a

```python
def generate_samples(m, var, n):
    return np.random.normal(m, np.sqrt(var), n)    

mu_prior, var_prior = 0, 4

mu_true, var_true = 7, 16

N = [1, 5, 10, 20, 50, 100, 1000]

for n in N:
    x = generate_samples(7, 16, n)

    mu_likelihood = np.sum(x) / n
    var_likelihood = var_true
    
    mu_posterior = (n * np.sum(x)/n * var_prior + var_likelihood * mu_prior) / (n * var_prior + var_likelihood)
    var_posterior = (var_prior * var_likelihood) / (n * var_prior + var_likelihood)
    
    # Print results
    print(f"N = {n}: Mean = {mu_posterior:.2f}, Variance = {var_posterior:.2f}")  
```

As the number of obsercations `N` is increasing, the mean of the _posterior_ function approximates the mean of the distribution N(7, 16), while the variance of the _posterior_ approximates zero, reducing the uncertainty about the estimation of the __mean__.

3.2.b

```python
for n in N:
    x = generate_samples(7, 16, n)

    mu_likelihood = np.sum(x) / n
    var_likelihood = var_true
    
    mu_posterior = (n * np.sum(x)/n * var_prior + var_likelihood * mu_prior) / (n * var_prior + var_likelihood)
    var_posterior = (var_prior * var_likelihood) / (n * var_prior + var_likelihood)
    
    x_axis = np.arange(-10, 25, 0.001)
    
    #define multiple normal distributions
    plt.plot(x_axis, norm.pdf(x_axis, mu_posterior, np.sqrt(var_posterior)), label='Posterior distribution')
    plt.plot(x_axis, norm.pdf(x_axis, mu_prior, np.sqrt(var_prior)), label='Prior distribution')
    plt.plot(x_axis, norm.pdf(x_axis, 7, np.sqrt(16)), label='Original distribution')
    plt.title('N = ' + str(n))

    #add legend to plot
    plt.legend()
    plt.show()
```
